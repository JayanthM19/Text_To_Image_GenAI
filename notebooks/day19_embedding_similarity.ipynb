```python
concepts = [
    "cat",
    "dog",
    "animal",
    "car",
    "vehicle"
]

for concept in concepts:
    print(concept)
```

```python
similarity_examples = {
    "cat": "animal",
    "dog": "animal",
    "car": "vehicle"
}

for word, related in similarity_examples.items():
    print(f"{word} ↔ {related}")
```

```python
import pandas as pd

data = {
    "Word 1": ["cat", "dog", "car"],
    "Word 2": ["animal", "animal", "vehicle"],
    "Similarity": [0.91, 0.89, 0.93]
}

df = pd.DataFrame(data)

print(df)
```
