```python
epochs = list(range(1, 11))

losses = [
    1.2,
    1.05,
    0.92,
    0.81,
    0.72,
    0.66,
    0.61,
    0.58,
    0.55,
    0.52
]

for epoch, loss in zip(epochs, losses):
    print(f"Epoch {epoch}: Loss = {loss}")
```

```python
import matplotlib.pyplot as plt

epochs = list(range(1, 11))

losses = [
    1.2,
    1.05,
    0.92,
    0.81,
    0.72,
    0.66,
    0.61,
    0.58,
    0.55,
    0.52
]

plt.figure(figsize=(8,5))

plt.plot(epochs, losses, marker="o")

plt.title("LoRA Training Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.grid(True)

plt.show()
```
