```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")

print("Pipeline loaded.")
```

```python
print(pipe.unet)
```

```python
total_params = sum(
    p.numel()
    for p in pipe.unet.parameters()
)

print(f"Total Parameters: {total_params:,}")
```

# How LoRA Works

LoRA injects small trainable matrices into attention layers.

Original weights remain frozen.

Only lightweight adapters are trained.

This enables:
- efficient training
- low VRAM usage
- modular fine-tuning

# DreamBooth Concept

DreamBooth fine-tunes diffusion models using a small custom image dataset.

Example:
- custom person
- product
- art style

LoRA can be combined with DreamBooth for lightweight personalization.
